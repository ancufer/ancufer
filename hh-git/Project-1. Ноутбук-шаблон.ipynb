3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…
    * Учитывайте эту особенность в вашем коде

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
# Функция принимает строку с опытом работы
# Выделяет в переменную exp_words список слов, первые два слова в списке [0](Опыт) и [1](работы)
# Затем идет срок работы
# Определяется как написан срок работы и возвращается кол-во месяцев работы

def get_experience(arg):
    
    if (pd.isna(arg)) or (arg == 'Не указано'):
        return np.nan

    exp_words = arg.split()[2:] # Сразу отсекаем элементы [0](Опыт) и [1](работы)

    len_exp_words = len(exp_words) # Определяем список, для опыта работы требуется не более 4 элементов например: 9 лет 9 месяцев   
    if len_exp_words > 3:
        exp_words = exp_words[:4]
               
    if ('год' in exp_words[1]) or (exp_words[1] == 'лет'): # Возможные опыты работы: 1 год, 3 года, 5 лет

        if len_exp_words > 3: # Проверка есть в списке индекс слова, которое может указывать на месяцы после указания лет
            if 'месяц' in exp_words[3]:
                return int(exp_words[0])*12+int(exp_words[2])
        return int(exp_words[0])*12 # Число слов меньше трех и второе слово год/лет, значит возвращаем просто кол-во лет x 12

    elif 'месяц' in exp_words[1]: # Если в первом слове только месяц, то возвращаем просто кол-во месяцев
        return int(exp_words[0])

    

hh_data['Опыт работы (месяц)'] = hh_data['Опыт работы'].apply(lambda x : get_experience(x))
hh_data = hh_data.drop(['Опыт работы'],axis=1)

print(hh_data['Опыт работы (месяц)'].median())
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск',
                  'Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']

# Так как в дальнейшем проверка осуществляется только по верхнему регистру
# переводим список городов в верхний регистр
million_cities = list(map(lambda x: x.upper(), million_cities))

# Определение типа города
# Аргумент передаётся в верхнем регистре (ЗАГЛАВНЫМИ буквами)
def city(arg):
    arg = arg.strip() # Убираем возможные пробелы справа и слева
    if 'МОСКВА' in arg:
        return 'Москва'
    elif 'ПЕТЕРБУРГ' in arg:
        return 'Санкт-Петербург'
    elif arg in million_cities:
        return 'город-миллионник'
    else:
        return 'другие'

# Одинаковая функция для признаков переезда и командировки
# Аргумент передаётся в верхнем регистре (ЗАГЛАВНЫМИ буквами)
# Если первое слово НЕ, значит не готов, не хочет, не желает и пр.
trip_move_ready = lambda arg: arg.strip().split()[0] != 'НЕ'

# Решил сделать так: в цикле пройтись по всем строкам в столбце ['Город, переезд, командировки']
# И добавлять в соответсвующие признакам списки значения
# Затем эти списки конвертировать в соответсвующие признакам столбцы
list_city = []
list_move = []
list_trip = []

for words in hh_data['Город, переезд, командировки']:
    words = words.upper()
    list_words = words.split(',')
    list_city.append(city(list_words[0]))
    if 'КОМАНДИРОВК' not in words:
        list_trip.append(1==2) # Так решил добавлять False в случае отсутствия информации о командировках
    if 'ПЕРЕЕ' not in words:
        list_move.append(1==2) # Так решил добавлять False в случае отсутствия информации о переезде
    
    for word in list_words[1:]:
        if 'ПЕРЕЕ' in word: # Переезд, переехать и прочее
            list_move.append(trip_move_ready(word))            
        elif 'КОМАНДИРОВК' in word:
            list_trip.append(trip_move_ready(word))
            

hh_data['Город'] = pd.DataFrame(list_city)
hh_data['Готовность к переезду'] = pd.DataFrame(list_move)
hh_data['Готовность к командировкам'] = pd.DataFrame(list_trip)
hh_data = hh_data.drop(['Город, переезд, командировки'],axis=1)

display(hh_data['Город'].value_counts(normalize=True))

mask = hh_data['Готовность к переезду'] & hh_data['Готовность к командировкам']

display(hh_data[mask]['Готовность к переезду'].sum() / hh_data.shape[0])
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
# Создание и удаление временных признаков
# для имеющегося df функция при выбранном столбце
# возвращает столбцы, с наименованиями соответствующим уникальным значениям
# выбранного столбца с простановкой True (если значение есть) и False (если значения нет)

def col_hot_encode(hh_data, col, sep=","):
    df = hh_data.copy() 
    # так может быть указано много типов занятостей/график работы через sep(в нашем случае запятая)
    # то в копии df создаём столбец, где каждый элемент это список из видов занятостей/графиков работы
    df[col + "_split"] = df[col].str.split(sep)

    # создаём список из всех видов занятостей/графиков работы
    list_el = [] 
    for el in df[col + "_split"] :
        list_el.extend(el)
        
    list_el = list(map(str.strip, list_el)) # убираем возможные лишние пробелы для каждого элемента 
    unique_values= set(list_el) # создаём уникальный список элементов (видов занятостей/графиков работы)
    # для каждого уникального элемента создаём в основном df столбец
    for uv in unique_values:
        hh_data[uv] = hh_data[col].str.contains(uv) # True (если значение есть) и False (если значения нет)

col_hot_encode(hh_data, 'Занятость')
col_hot_encode(hh_data, 'График')
        
mask =  hh_data['проектная работа'] & hh_data['волонтерство']

display(sum(hh_data[mask]['проектная работа']))

mask = hh_data['вахтовый метод'] & hh_data['гибкий график']

display(sum(hh_data[mask]['вахтовый метод']))

hh_data = hh_data.drop(columns=['Занятость', 'График'])
```

Осталось только понять, откуда брать дату, по которой определяется курс? А вот же она - в признаке **"Обновление резюме"**, в нем содержится дата и время, когда соискатель выложил текущий вариант своего резюме. Нас интересует только дата, по ней бы и будем сопоставлять курсы валют.

Теперь у нас есть вся необходимая информация для того, чтобы создать признак "ЗП (руб)" - заработная плата в рублях.

После ответа на контрольные вопросы удалите исходный столбец заработной платы "ЗП" и все промежуточные столбцы, если вы их создавали.

Итак, давайте обсудим возможный алгоритм преобразования: 
1. Перевести признак "Обновление резюме" из таблицы с резюме в формат datetime и достать из него дату. В тот же формат привести признак "date" из таблицы с валютами.
2. Выделить из столбца "ЗП" сумму желаемой заработной платы и наименование валюты, в которой она исчисляется. Наименование валюты перевести в стандарт ISO согласно с таблицей выше.
3. Присоединить к таблице с резюме таблицу с курсами по столбцам с датой и названием валюты (подумайте, какой тип объединения надо выбрать, чтобы в таблице с резюме сохранились данные о заработной плате, изначально представленной в рублях). Значение close для рубля заполнить единицей 1 (курс рубля самого к себе)
4. Умножить сумму желаемой заработной платы на присоединенный курс валюты (close) и разделить на пропорцию (обратите внимание на пропуски после объединения в этих столбцах), результат занести в новый столбец "ЗП (руб)".


```python
# Создание dataframe из файла с валютами и их признаками
currencies = pd.read_csv('data/ExchangeRates/ExchangeRates.csv',sep=',')
```

```python
# Конвертация в формат datetime признака date в таблице валют 
currencies['date'] = pd.to_datetime(currencies['date'] , format="%d/%m/%y").dt.date
```

```python
# Приведение к типу datetime признака 'Обновление резюме'
# Создание нового признака с отдельно выделенной датой
hh_data['Обновление резюме'] = pd.to_datetime(hh_data['Обновление резюме'] , format="%d.%m.%Y %H:%M").dt.date
```

```python
# Вводим словарь перевода названия валют в ISO наименование
# Для избежания возможных неточностей все ключи словаря заданы с ЗАГЛАВНОЙ буквы
dict_hh_curr = {
    "ГРН" :	"UAH",
    "USD" : "USD",
    "EUR" :	"EUR",
    'БЕЛРУБ' : 'BYN',
    'KGS' :	'KGS',
    'СУМ' : 'UZS',
    'AZN' : 'AZN',
    'KZT' : 'KZT',
    'РУБ' : "РУБ"
    }
```

```python
# создаём временный столбец из списков, у которых первый элемент зарплаты, второй наменованием валюты
hh_data['ЗП_сплит'] = hh_data['ЗП'].str.split()
hh_data['ЗП (руб)'] = hh_data['ЗП_сплит'].str[0] # первый элемент это сумма
# второй элемент это валюта убираем пробелы, точки и переводим в заглавные буквы (как в словаре)
hh_data['Валюта'] = hh_data['ЗП_сплит'].str[1]. \
    str.strip().\
    str.replace(".", "").\
    str.upper()
    
hh_data.drop(columns='ЗП_сплит', inplace=True) # удаляем уже ненужный столбец

hh_data['Валюта'] = hh_data['Валюта'].map(dict_hh_curr) # Перевод валюты в ISO наименование
```

```python
# Добавление через объединение по валюте и дате расчета её курса признаков валюты

hh_data = hh_data.merge(    
    currencies,
    how='left',
    left_on=['Обновление резюме','Валюта'],
    right_on=['date','currency']
    )

# Так как в таблице с валютами нет рублей, то для рублей признаки курса и пропорции задаются "вручную"
mask = hh_data['Валюта'] == "РУБ"
hh_data.loc[mask, 'proportion'] = 1.0
hh_data.loc[mask, 'close'] = 1.0
hh_data['ЗП (руб)'] = hh_data['ЗП (руб)'].astype(float)

```

```python
# Создание нового столбца и удаление всех промежуточных столбцов
hh_data['ЗП (руб)'] = hh_data['ЗП (руб)'] * hh_data['close'] / hh_data['proportion']

hh_data = hh_data.drop(columns=['ЗП', 'Валюта', 'currency','per','date','time','close','vol','proportion'])

display(hh_data['ЗП (руб)'].median())
```

```python
# Функция Skillfactory
def outliers_z_score(data, feature, log_scale=False,left=3,right=3):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

# "Послабление" вправо на 1 сигму
outliers, cleaned = outliers_z_score(hh_data, 'Возраст', True,3,4)
print(f'Число выбросов по методу z-отклонения: {outliers.shape[0]}')
print(f'Результирующее число записей: {cleaned.shape[0]}')
```

График логарифма возраста построенный по методу 3-х сигм ($\sigma$) имеет правостороннюю ассимметрию. Линия среднего значения (красная) находится правее моды. На графике показано увеличение правой границы на одну сигму ($\sigma$) (*вертикальная зелёная пунктирная линия*)
